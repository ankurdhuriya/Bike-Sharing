### Data Preparation for Model Training

#### Dummy variable creation for month, weekday, weathersit and season variables.

```python
months_df=pd.get_dummies(df.month,drop_first=True, dtype=int)
weekdays_df=pd.get_dummies(df.weekday,drop_first=True, dtype=int)
weathersit_df=pd.get_dummies(df.weathersit,drop_first=True, dtype=int)
seasons_df=pd.get_dummies(df.season,drop_first=True, dtype=int)
```

```python
df_new = pd.concat([df,months_df,weekdays_df,weathersit_df,seasons_df],axis=1)
```

```python
df_new.head(4)
```

```python
df_new.shape
```

dropping unnecessary columns as we have already created dummy variable out of it.

```python
df_new.drop(['season','month','weekday','weathersit'], axis = 1, inplace = True)
```

```python
df_new.head(4)
```

#### Splitting the data

```python
np.random.seed(121)
df_train, df_test = train_test_split(df_new, train_size = 0.7, random_state = 2023)
```

```python
df_train.shape, df_test.shape
```

Using MinMaxScaler to Rescaling the features

```python
scaler = MinMaxScaler()
```

Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

```python
num_vars = ['temp','atemp','humidity','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head(2)
```

```python
plt.figure(figsize = (25,25))
matrix = np.triu(df_train.corr())
sns.heatmap(df_train.corr(), annot = True, cmap="RdYlGn", mask=matrix)
plt.show()
```

cnt seems to have correlation with year variable and temp. Similarly, Misty and humidity show correlation. Spring season with Jan and Feb month, Summer season with may month and Winter season with oct and nov month show good correlation.

### Building Model

```python
# Building the Linear Model

y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Recursive feature elimination 

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
#List of variables selected in top 15 list

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# selecting the selected variable via RFE in col list

col = X_train.columns[rfe.support_]
print(col)
```

```python
# checking which columns has been rejected

X_train.columns[~rfe.support_]
```

```python
# Generic function to calculate VIF of variables

def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# calculate VIF

calculateVIF(X_train_rfe)
```

```python
# As temp , workingday and atemp shows high VIF values hence we can drop it
X_train_new = X_train_rfe.drop(['temp', 'atemp', 'workingday'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

#### Building 1st linear regression model

```python
X_train_lm_1 = sm.add_constant(X_train_new)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
```

```python
print(lr_1.summary())
```

```python
# We can drop humidity, july, sat and sun variable as it has high p-value
X_train_new = X_train_new.drop(['humidity', 'july', 'sat', 'sun'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

VIF values seems to be good now. But we will see if we can reduce further ?

#### Building 2nd linear regression model

```python
X_train_lm_2 = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

We can cosider the above model i.e lr_2, as it seems to have very low multicolinearity between the predictors and the p-values for all the predictors seems to be significant. </br>
F-Statistics value of 194.2 (which is greater than 1) and the p-value of 3.46e-148 i.e almost equals to zero, states that the overall model is significant

```python
# Checking the parameters and their coefficient values
lr_2.params
```

### Residual Analysis of the train data and validation

```python
X_train_lm_2
```

```python
y_train_pred = lr_2.predict(X_train_lm_2)
```

#### Normality of error terms

```python
# Plot the histogram of the error terms

fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20) 
plt.xlabel('Errors', fontsize = 18)
```

Error terms are following normal distribution

#### Multi Colinearity

```python
calculateVIF(X_train_new)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(X_train_new.corr(),annot = True, cmap="RdYlGn")
plt.show()
```

VIF values are less than 3 which is good and also there is no multicolinearity as seen from the heatmap.

#### Homoscedasticity

```python
y_train_pred = lr_2.predict(X_train_lm_2)
residual = y_train - y_train_pred
sns.scatterplot(x=y_train,y=residual)
plt.plot(y_train,(y_train - y_train), '-r')
plt.xlabel('Count')
plt.ylabel('Residual')
plt.show()
```

No visible pattern observed from above plot for residuals.

Independence of residuals </br>
Durbin-Watson value of final model lr_2 is 2.011, which signifies there is no autocorrelation

### Making Predictions Using the Final Model

Now that we have fitted the model and checked the normality of error terms, it's time to go ahead and make predictions using the final, i.e. 2nd model.

```python
# Applying scaling on the test dataset

num_vars = ['temp', 'atemp', 'humidity', 'windspeed','cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
col1 = X_train_new.columns

X_test = X_test[col1]

# Adding constant variable to test dataframe
X_test_lm_2 = sm.add_constant(X_test)
```

```python
y_pred = lr_2.predict(X_test_lm_2)
```

```python
r2 = r2_score(y_test, y_pred)
round(r2,4)
```

### Model Evaluation

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20) 
plt.xlabel('y_test', fontsize = 18)
plt.ylabel('y_pred', fontsize = 16) 
```

```python
round(lr_2.params,4)
```

```python
# Calculating Adjusted-R^2 value for the test dataset

adjusted_r2 = round(1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1),4)
print(adjusted_r2)
```

```python
# Visualizing the fit on the test data
# plotting a Regression plot

plt.figure()
sns.regplot(x=y_test, y=y_pred, ci=68, fit_reg=True,scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)
plt.show()
```

Comparision between Training and Testing dataset: </br>
- Train dataset R^2          : 0.756
- Test dataset R^2           : 0.7427
- Train dataset Adjusted R^2 : 0.752
- Test dataset Adjusted R^2  : 0.7329
